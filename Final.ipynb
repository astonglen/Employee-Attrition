```python
#Creating a list with the names of continous features
num_cols = []
for col in df.columns:
    if col not in cat_cols:
        num_cols.append(col)

#Removing Attribution as it is a label
num_cols.remove('Attrition')
```

```python
#Plotting box plots for checking outliers
for col in num_cols:
    sns.boxplot(df[col])
    plt.show()
```

```python
def change_to_num(val):
    if val=='Yes':
        return 1
    else :
        return 0
    
```

```python
#Converting the label from string to numerical
df['Attrition'] = df['Attrition'].apply(change_to_num)
```

```python
#One hot encoding all the categorical features using pandas .get_dummies
for col in cat_cols:
    df[col] = df[col].astype(str)
    df = pd.concat([df,pd.get_dummies(df[col], prefix=col)],axis=1)

```

```python
#Removing the original catrgorical column
for col in cat_cols:
    df.drop(col,axis=1,inplace=True)
```

```python
print(df.columns.values)
```

```python
#Plotting the distibution of each continous feature
for col in num_cols:
    
    sns.distplot(df[col])
    plt.show()
```

### Modeling 

#### Balancing the dataset

```python
#Using smote to balance the dataset
from imblearn.over_sampling import SMOTE
X = df.drop('Attrition',axis=1)
y = df['Attrition']
oversample = SMOTE()
X_over, y_over = oversample.fit_resample(X, y)
```

```python
print("Number of samples before SMOTE {}:".format(X.shape[0]))
print("Number of samples before SMOTE {}:".format(X_over.shape[0]))
```

```python
#creating a new dataframe which has oversampled data
df_over = pd.concat([X_over,y_over],axis=1)
df_over.shape
```

#### As none of the continous feature has a normal distribution we will be using "Min Max Scaler" instead of "Standard Scaler"

```python
#Min max scale the numerical columns
from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()

x_temp = df_over[num_cols].values
x_scaled = mm_scaler.fit_transform(x_temp)
df_temp = pd.DataFrame(x_scaled, columns=num_cols, index = df_over.index)
df_over[num_cols] = df_temp
```

```python
df_over.describe()
```

#### We can see that all the continous features are scaled between 0 and 1

```python
#Splitting the data into testing and training
from sklearn.model_selection import train_test_split
X = df_over.drop('Attrition',axis=1)
y = df_over['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                         y, 
                                                         train_size= 0.70,
                                                         random_state=0);
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
accuracy_list = []
model_list = []

#Function to print the confusion matrix, classification report 
def output_metrics(predictions):
    ''' 
    Predictions are the predicted values for the test data 
    '''
    print("-------Confusion Matrix--------")
    print(confusion_matrix(y_test,predictions))
    print('\n')
    print("-------Classification Report--------")
    print(classification_report(y_test,predictions))
    print("-------Accuracy Score--------")
    print(accuracy_score(y_test, predictions))
    accuracy_list.append(accuracy_score(y_test, predictions))
    
def train_model(model,X_train,y_train,X_test,y_test):
    '''
    model            :  Model which is going to be used 
    X_train ,y_train :  features and labels of training data
    X_test, y_test   :  features and labels of testing data
    '''
    model_list.append(model)
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    output_metrics(predictions)
```

### K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Knn with 1 nearest neighbour
knn = KNeighborsClassifier(n_neighbors=1)
train_model(knn, X_train, y_train, X_test, y_test)
```

```python
#Finding the optimal k value 
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
#Plotting the error rate for different k values
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
#lowest error rate with corresponding k value
print("The lowest error rate is {} for n_neighbours = {}".format(min(error_rate),error_rate.index(min(error_rate))+1))
```

```python
#KNN with 2 neighbours
knn2 = KNeighborsClassifier(n_neighbors=error_rate.index(min(error_rate))+1)
train_model(knn2, X_train, y_train, X_test, y_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600)
train_model(rfc, X_train, y_train, X_test, y_test)
```

```python
features = df.columns
importance = rfc.feature_importances_
indices = np.argsort(importance)
plt.figure(1,figsize=(10,20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importance[indices], color='lightblue', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.xlabel('Relative Importance')
```

### Support Vector Machines

```python
from sklearn.svm import SVC
svc_model = SVC()
train_model(svc_model, X_train, y_train, X_test, y_test)
```

```python
# Hyper parameter optimization for SVM using GridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
train_model(grid, X_train, y_train, X_test, y_test)
```

```python
grid_pred = grid.predict(X_test)
```

```python
output_metrics(grid_pred)
```
